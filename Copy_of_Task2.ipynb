Jupyter Notebook lets you to write line by line of code. Python script expects you to define an entire program consisting of classes and functions. Python script is a high level programming language and is basically plain code(a .py file).Jupyter Notebook primarily utilises is a .ipynb file which is essentially a python notebook which contains the notebook code, the execution results in a specific format.

Remove all the rows having type as 'Hex'

```python
df.drop(df[df.Type=='Hex'].index,inplace=True)
df
```

Now remove the rows from the dataframe having NaN values for column 'Light'.

```python
df.dropna()
```

Filter out the spells on the basis of their Type and make a bar graph showing your analysis of the number of spells and their distribution in the five types - Charm, Jinx, Transfiguration, Curse.

```python
freq={}
for Type in df.Type.unique():
  c=0
  for item in df.Type:
    if item==Type:
      c+=1
      freq[Type]=c
plt.figure(figsize=(20,7))
plt.bar(range(len(df.Type.unique())),list(freq.values()))
plt.xticks(range(len(freq)),df.Type.unique())
plt.show()
```
